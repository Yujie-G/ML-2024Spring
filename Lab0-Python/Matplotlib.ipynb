# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 6])
ypoints = np.array([0, 100])

plt.plot(xpoints, ypoints)
# plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,2*np.pi,0.1)  # start,stop,step
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y,x,z)
plt.show()
```

```python
x = np.arange(-2*np.pi,2*np.pi,0.1)  # start,stop,step
t = np.tanh(x)
plt.plot(x,t)
plt.show()
```
